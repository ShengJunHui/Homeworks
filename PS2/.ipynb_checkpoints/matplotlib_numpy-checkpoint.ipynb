```python
def chi_squared(m_given, hubble_data1, hubble_data2):
    m_givenlist=[]
    for i in range(len(hubble_data1)):
        m_givenlist.append(m_given)
    
    hubble_data3=[]   #divided by data1 and data2
    for i in range(len(hubble_data1)):
        hubble_data3.append(hubble_data1[i]/ hubble_data2[i])
    hubble_data3=  hubble_data3- m_givenlist
    np.power(hubble_data3, 2)
    return sum(hubble_data3)
```

## Problem 3) Intro to Numerical Integration (10 Points)

To preview next week's lesson, lets play around with some basic numerical integration techniques. Consider the following function;

$$E(x) = \frac{2}{\sqrt{\pi}}\int_{0}^{x}e^{-t^{2}}dt.$$

The integral within this function can not be done analytically; i.e. it doesn't possess a nice functional solution like integrals encountered in the standard calculus course. Instead we will have to use numerical integration techniques to solve for it.

1) Code up a function for $E(x)$, you may use the basic trapizoidal integration solver below. Plot the function for values between $x=1$ and $x=100$.

```python
import numpy as np
import matplotlib.pylab as plt

def func(x):
    return (2/ np.sqrt(np.pi))* np.exp(-x**2)

def trap_int(f, a, b, N):
    """
    very simple trapezoidal integration solver
    """
    # bin width
    h = (b-a)/float(N)
    
    # get x-values
    x_range = np.linspace(a, b, N+1)
    
    # get y-values
    y_range = f(x_range)
    
    # evaluate trapezoidal sum
    I = h*(0.5*y_range[0] + 0.5*y_range[-1] + sum(y_range[1:-1]))
    
    return I

x= np.linspace(1, 100,10000)
I_list=[]
for i in range(10000):
    I_list.append( trap_int(func, 0, x[i], 1000))
plt.plot(x,I_list,label="integration")
plt.grid()
plt.xlabel("x",fontsize=16)
plt.ylabel("I",fontsize=16)
plt.legend()
```

2) What is the minimum number of slices (i.e. what value of $N$) is needed for the following accuracy?

$$E(0.5) = 0.52049987(7)$$
$$E(1.0) = 0.84270079(2)$$
$$E(1.5) =  0.96610514(6)$$

```python
def Slices(N, trap_int, b, E_exact):
    while abs(trap_int(func, 0, b, N)- E_exact) >= 0.00000001:
        N= N+1
    return N
```

```python
Slices(1, trap_int, 0.5, 0.52049987)
```

```python
Slices(1, trap_int, 1.0, 0.84270079)
```

```python
Slices(1, trap_int, 1.5, 0.96610514)
```

Later on in the course we will learn more about running simulations of particles. In this excersise we will code up a very simple n-body simulation meant to show the formation of structure. We will assume the particles move only under gravitational interactions.

$$F_G = k m_1 m_2 /r^2$$

1) At early time it is thought that particles form at random positions and velocities from an underlying field. Generate 1,000 particles at random positions with random velocities in two dimensions. Assume a box size with units between 0 and 1. Have velocities be in each direction have a Gaussian distribution with mean 0 and standard deviation 0.01. Plot the field!

```python
import numpy as np
import matplotlib.pylab as plt
 
x= np.random.random(1000)
y= np.random.random(1000)
u= np.random.normal(0, 0.01, 1000) 
v= np.random.normal(0, 0.01, 1000) 

# plotting the field: (x,y) is the location; (u,v) is the velocity direction 
plt.quiver(x, y, u, v)
plt.title("Field within the box")
```

2) Write a function to calculate the force on each particle. Remember that force is a vector quanity.

```python
# in order to simplify the question, assume that parameter k, m1 and m2 is one 
x_new= x.reshape(1000,1)
y_new= y.reshape(1000,1)
r= np.hstack([x_new, y_new]) # information of position of each particle
force= np.zeros([1000,2])

def Force(r, force):

    for i in range(0,1000):
        force_total_x= 0
        force_total_y= 0
        for j in range(0,i):
            force_total_x += (r[j,0]- r[i,0])/ ((r[j,0]-r[i,0])**2+(r[j,1]-r[i,1])**2)**(2/3)
            force_total_y += (r[j,1]- r[i,1])/ ((r[j,0]-r[i,0])**2+(r[j,1]-r[i,1])**2)**(2/3)
        for k in range(i+1,1000):
            force_total_x += (r[k,0]- r[i,0])/ ((r[k,0]-r[i,0])**2+(r[k,1]-r[i,1])**2)**(2/3)
            force_total_y += (r[k,1]- r[i,1])/ ((r[k,0]-r[i,0])**2+(r[k,1]-r[i,1])**2)**(2/3)    
        force[i,:]= [force_total_x, force_total_y]
        # force list contains the information of the force each particle suffers
    
    return force

Force(r, force)
```
